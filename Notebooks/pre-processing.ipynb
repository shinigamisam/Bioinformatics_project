```python
unique_values_dict = {}
columns_to_discard = []  # List to store columns with few unique values

# Define threshold for "few" unique values
max_unique_values = 10  # Adjust this threshold as needed (e.g., 2, 3, 10)

# Loop through each column to get unique values
for column in df.columns:
    unique_values = df[column].unique()
    unique_count = len(unique_values)
    
    # Store unique values and count in the dictionary
    unique_values_dict[column] = {
        'unique_values': unique_values,
        'count': unique_count
    }
    
    # Flag columns with <= max_unique_values for discarding
    if unique_count <= max_unique_values:
        columns_to_discard.append(column)

# Print unique values for columns with <= max_unique_values
print(f"\nColumns with {max_unique_values} or fewer unique values:")
val_col_1 = 0
for column, info in unique_values_dict.items():
    if info['count'] <= max_unique_values:
        print(f"Column: {column}, Unique values: {info['unique_values']}, Count: {info['count']}")
        val_col_1 += 1  # Increment counter for each column with few unique values

# Print the total number of columns with few unique values
print(f"Total number of columns with {max_unique_values} or fewer unique values: {val_col_1}")
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.feature_selection import VarianceThreshold
import time

# Start timing
start_time = time.time()

# Step 1: Load the dataset
df = pd.read_csv("../Dataframes/fixed_merge1.csv")
print("Original shape:", df.shape)

# Step 2: Separate features and labels
label_columns = ['Unnamed: 0', 'patient_id', 'cancer_type', 'mutational_subclass', 'cancer']
feature_columns = [col for col in df.columns if col not in label_columns]
X = df[feature_columns]  # Gene expression features
y = df['cancer']    # Labels
misc = df[["Unnamed: 0", "mutational_subclass", "cancer_type"]]
print("Feature columns:", len(feature_columns))

# Step 3: Normalize the data
# Log-transform (for raw counts or skewed data)
X = np.log1p(X)
# OR Standardize (uncomment if preferred)
# scaler = StandardScaler()
# X = pd.DataFrame(scaler.fit_transform(X), columns=X.columns)

# Step 4: Quick variance filter to reduce columns
selector = VarianceThreshold(threshold=0.01)  # Remove low-variance genes
X_transformed = selector.fit_transform(X)
selected_columns = [feature_columns[i] for i in range(len(feature_columns)) if selector.get_support()[i]]
X = pd.DataFrame(X_transformed, columns=selected_columns)
print("Shape after variance filter:", X.shape)

# Step 5: Correlation filter with batch processing
chunk_size = 5000  # Adjust based on system memory
to_drop = []
for i in range(0, len(X.columns), chunk_size):
    print(f"Processing chunk {i//chunk_size + 1}...")
    chunk = X.iloc[:, i:i+chunk_size]
    corr_matrix = chunk.corr().abs()
    upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))
    #chunk_drop = [chunk.columns[j] for j in range(chunk.shape[1]) if any(upper.iloc[:, j] < 0.02)]
    chunk_drop = [chunk.columns[j] for j in range(chunk.shape[1]) if any((upper.iloc[:, j] > 0.2) | (upper.iloc[:, j] < -0.2))]

    to_drop.extend(chunk_drop)
to_drop = list(set(to_drop))  # Remove duplicates
print(f"Columns to drop: {len(to_drop)}")


# Drop highly correlated columns
X_reduced = X.drop(columns=to_drop)
print("Reduced feature shape:", X_reduced.shape)

# Step 6: Combine reduced features with labels
df_reduced = pd.concat([y, X_reduced, misc], axis=1)
print("Reduced dataset shape:", df_reduced.shape)

# Save the reduced dataset
# Drop the substring 'cancer type' from the 'cancer_type' column
df_reduced['cancer_type'] = df_reduced['cancer_type'].str.replace('cancer type:', '')

# Drop the substring 'mutational subclass' from the 'mutational_subclass' column
df_reduced['mutational_subclass'] = df_reduced['mutational_subclass'].str.replace('mutational subclass:', '')

# Drop the substring 'patient id:' from the 'patient_id' column

df_reduced.to_csv("../Dataframes/reduced_merge1_corr.csv", index=False)
print("Reduced dataset saved as '../Dataframes/reduced_merge1_corr.csv'")

# Print runtime
print(f"Total runtime: {(time.time() - start_time)/60:.2f} minutes")

# Optional: Basic validation with a model (e.g., for 'cancer' binary classification)
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
y_encoded = le.fit_transform(y)  # Or 'cancer_type' for multiclass
model = RandomForestClassifier(random_state=42)
scores = cross_val_score(model, X_reduced, y_encoded, cv=5, scoring='accuracy')
print("Mean CV Accuracy (on reduced data):", scores.mean())
df.to_csv('../Dataframes/reduced_via_correl.csv', index=False)
```
