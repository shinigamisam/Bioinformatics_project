```python
series_df2["unique_id"] = series_df2["unique_id"].astype(str).str.replace('"', '')
series_df2["patient_id"] = series_df2["patient_id"].astype(str).str.replace('"', '').str.strip()
series_df2["cancer_type"] = series_df2["cancer_type"].astype(str).str.replace('"', '').str.strip()
series_df2["mutational_subclass"] = series_df2["mutational_subclass"].astype(str).str.replace('"', '').str.strip()
```
